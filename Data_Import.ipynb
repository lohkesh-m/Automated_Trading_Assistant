```python
import yfinance as yf

import pandas as pd
from pandas_datareader import data as pdr

from datetime import datetime
```

```python
Watchlist_tickers=pd.read_csv('Data/Watchlist_yahoo_tickers.csv')
Watchlist_list=Watchlist_tickers['Tickers'].tolist()
end = datetime.today()
begin=end-pd.DateOffset(5*365)
st=begin.strftime('%Y-%m-%d') 
ed=end.strftime('%Y-%m-%d')
data=[]
niftyd_list=[]
for i,k in enumerate(Watchlist_list):
    print(f"At iteration {i}")
    try:
        data.append(pdr.get_data_yahoo(k,st,ed))
        niftyd_list.append(k)
        
    except Exception:
        print('Not found',k)
        Watchlist_list.remove(k)
        pass
    
Watchlist_tickers=pd.DataFrame(Watchlist_list,columns=['Tickers'])
Watchlist_tickers.to_csv('Data/Watchlist_yahoo_tickers.csv')

writer=pd.ExcelWriter("Data/Backtest_Data.xlsx")
for i, A in enumerate(data):
    A.to_excel(writer,sheet_name="{0}".format(niftyd_list[i]))
writer.save()
print("0")
```
